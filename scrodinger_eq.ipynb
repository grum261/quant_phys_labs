# 2. Формирование матрицы Гамильтона

#### Вводим размер матрицы

```python
N = 1001
```

#### Формируем и выводим матрицу NxN



```python
x = np.linspace(-1, 1, N)
```

```python
T = np.zeros((N - 2, N - 2))
for k in range(-1, 2):
    row_idx, col_idx = [np.diag(axis, k) for axis in np.indices(T.shape)]
    if k != 0:
        T[row_idx, col_idx] = 1
    else:
        T[row_idx, col_idx] = -2
```

```python
V = np.diag(x[1:-1]**4)
V
```

```python
H = -T + V
H
```

# 3. Находим собственные значения и соответствующие функции

```python
eig, eve = np.linalg.eigh(H)
```

#### Выводим 20 первых значений

```python
eig[:20]
```

#### Значения энергий

```python
En = eig/(eig[1] - eig[0])
En[:4]
```

# 4. Графики

## 4.1 График функции $\psi$ (x) * $\psi$ (x)

```python
plt.figure(figsize=(15, 10))
for n in range(4):
    plt.plot(x[1:-1], eve[:, n]*eve[:, n], lw=3, label=f'{n}')
    plt.xticks(np.arange(-1, 1.5, 0.25))
    plt.xlabel('x', size=15)
    plt.ylabel('$\psi$ (x) * $\psi$ (x)', size=15)
plt.legend(fontsize=15)
plt.grid()
plt.axhline(color='k')
plt.axvline(color='k')
```

## 4.2 График функции $\psi$ (x)

```python
plt.figure(figsize=(15, 10))
for n in range(4):
    plt.plot(x[1:-1], eve[:, n], lw=3, label=f'{n}')
    plt.xticks(np.arange(-1, 1.5, 0.25))
    plt.xlabel('x', size=15)
    plt.ylabel('$\psi$ (x)', size=15)
plt.legend(fontsize=15)
plt.grid()
plt.axhline(color='k')
plt.axvline(color='k')
```
